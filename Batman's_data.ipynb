```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://www.espncricinfo.com/series/men-s-t20-asia-cup-2022-1327237/afghanistan-vs-sri-lanka-1st-match-group-b-1327269/full-scorecard'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find('table', class_ = 'ds-w-full ds-table ds-table-md ds-table-auto ci-scorecard-table')
```

```python
t1 = soup.find('table', class_ = 'ds-w-full ds-table ds-table-md ds-table-auto ci-scorecard-table')
```

```python
t1
```

```python
th = t1.find_all('th')
```

```python
th
```

```python
head = [th.text.strip() for th in th]
```

```python
head
```

```python
head[1] = 'Description'
```

```python
head
```

```python
df = pd.DataFrame(columns = head)
```

```python
df
```

```python
data = t1.find('tbody')
data
```

```python
batmans_data = data.find_all('tr', class_ = lambda x: x != 'ds-hidden')
```

```python
batmans_data[:-3]
```

```python
for row in batmans_data[:-3]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
